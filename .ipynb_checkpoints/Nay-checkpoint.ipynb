```python
import numpy as np

 #https://www.kaggle.com/code/andresjejen/regresi-n-logistica, ver documentacion
    
def kernel_sigmoid(z):
    '''
   
    Parameters
    ----------
    z (float or np.array): Valor o array de valores a los que se aplicará la función sigmoide.
    
    Returns
    ------
    (float or np.array): Resultado de aplicar la función sigmoide a z.
    '''
    return 1 / (1 + np.exp(-z))
    

def forward_propagation(W, b, X, y):
    '''Implemente la función de coste y su gradiente para la propagación
    
    Parameters
    ----------
    W (np.array): Pesos del modelo.
    b (float): Sesgo 
    X (np.array): Conjunto de características.
    y (np.array): Etiquetas vector
    
    Returns
    ------
    A (np.array): Activaciones después de aplicar la función sigmoide.
    cost (float): coste de log-verosimilitud para la regresión logística
    '''
    # Número de datos
    data_number = X.shape[0]
    
    Z = np.dot(W, X.T) + b
    # Aplicación de la función sigmoide
    A = kernel_sigmoid(Z)
    # Cálculo del costo
    cost = (-1 / data_number) * np.sum(y * np.log(A) + (1 - y) * np.log(1 - A))

    return A, cost

def backward_propagation(X, A, y):
    '''Calcula los gradientes de los pesos y el sesgo.
    
    Parameters
    ----------
    X (np.array): Conjunto de características.
    A (np.array): Activaciones después de aplicar la función sigmoide.
    y (np.array): Etiquetas vector
    
    Returns
    ------
    dW (np.array): Gradiente con respecto a los pesos.
    db (float): Gradiente con respecto al sesgo.
    '''
    # Número de datos
    data_number = X.shape[0]
    
    # Gradiente con respecto a W
    dW = (1 / data_number) * np.dot((A - y), X)
    
    # Gradiente con respecto a b
    db = (1 / data_number) * np.sum(A - y)

    return dW, db

def optimize(W, b, X, y, num_iterations, learning_rate):
    '''Optimiza los parámetros usando el descenso de gradiente.
    
    Parameters
    ----------
    W (np.array): Pesos del modelo.
    b (float): Sesgo del modelo.
    X (np.array): Conjunto de características.
    y (np.array): Etiquetas vector
    num_iterations (int): Número de iteraciones para el descenso de gradiente.
    learning_rate (float): Tasa de aprendizaje.
    
    Returns
    ------
    params (dict): Diccionario con los parámetros optimizados (W y b).
    gradients (dict): Diccionario con los gradientes calculados (dW y db).
    costs (list): Lista con los costos calculados durante el proceso de optimización.
    '''
    # Almacenar los costos
    costs = []

    for i in range(num_iterations):
        # Propagación hacia adelante
        A, cost = forward_propagation(W, b, X, y)
        # Propagación hacia atrás
        dW, db = backward_propagation(X, A, y)

        # Actualizar parametros
        W = W - learning_rate * dW
        b = b - learning_rate * db

        # Costo registro
        if i % 100 == 0:
            costs.append(cost)

    # Diccionario con los parámetros optimizados
    params = {
        "W": W,
        "b": b
    }

    # Diccionario con los gradientes
    gradients = {
        "dW": dW,
        "db": db
    }

    return params, gradients, costs

def predict(W, b, X):
    '''Realiza predicciones usando el modelo entrenado.
    
    Parameters
    ----------
    W (np.array): Pesos del modelo.
    b (float): Sesgo del modelo.
    X (np.array): Conjunto de características.
    
    Returns
    ------
    y_prediction (np.array): Predicciones del modelo para el conjunto de características X.
    '''
    data_number = X.shape[0]
    # Inicialización de la predicción
    y_prediction = np.zeros((1, data_number))

    Z = np.dot(W, X.T) + b
    # Aplicación de la función sigmoide
    A = kernel_sigmoid(Z)

    for i in range(A.shape[1]):
        y_prediction[0, i] = 1 if A[0, i] > 0.5 else 0

    return y_prediction

def model_regresion_logistico(X_train, y_train, X_val, y_val, num_iterations=2000, learning_rate=0.5):
    '''Entrena el modelo de regresión logística y evalúa su rendimiento.
    
    Parameters
    ----------
    X_train (np.array): Conjunto de entrenamiento de características.
    y_train (np.array): Etiquetas verdaderas para el conjunto de entrenamiento.
    X_val (np.array): Conjunto de validación de características.
    y_val (np.array): Etiquetas verdaderas para el conjunto de validación.
    num_iterations (int): Número de iteraciones para el descenso de gradiente ( 2000).
    learning_rate (float): Tasa de aprendizaje (0.5)
    
    Returns
    ------
    lista (dict): Diccionario con la precisión en entrenamiento y validación, y los costos durante el proceso de entrenamiento.
    '''
    # Dimensiones conjunto de entrenamiento
    dimensions = X_train.shape[1]
    # Inicializar parametros
    W = np.zeros(shape=(1, dimensions))
    b = 0

    # Optimiza de los parámetros
    params, gradients, costs = optimize(W, b, X_train, y_train, num_iterations, learning_rate)
    W = params["W"]
    b = params["b"]

    # Predicción en el conjunto de validación
    y_prediction_validation = predict(W, b, X_val)
    
    # Predicción en el conjunto de entrenamiento
    y_prediction_train = predict(W, b, X_train)

    # Resultados
    lista = {
        "Ajuste entrenamiento": 100 - np.mean(np.abs(y_prediction_train - y_train)) * 100,
        "Ajuste testeo": 100 - np.mean(np.abs(y_prediction_validation - y_val)) * 100,
        "Costo en proceso": costs
    }

    return lista
```

# Optimizar

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import time

# Funciones para el modelo optimizado
def kernel_sigmoid(z):
    return 1 / (1 + np.exp(-z))

def forward_propagation(W, b, X, y):
    data_number = X.shape[0]
    Z = np.dot(W, X.T) + b
    A = kernel_sigmoid(Z)
    cost = (-1 / data_number) * np.sum(y * np.log(A) + (1 - y) * np.log(1 - A))
    return A, cost

def backward_propagation(X, A, y):
    data_number = X.shape[0]
    dW = (1 / data_number) * np.dot((A - y), X)
    db = (1 / data_number) * np.sum(A - y)
    return dW, db



def optimize(W, b, X, y, num_iterations, learning_rate):
    costs = []
    for i in range(num_iterations):
        A, cost = forward_propagation(W, b, X, y)
        dW, db = backward_propagation(X, A, y)
        W -= learning_rate * dW # Cambio: recordar que escribir a=a-b es igual a  escribir a-=b
        b -= learning_rate * db
        if i % 100 == 0:
            costs.append(cost)
    params = {
        "W": W,
        "b": b
    }
    gradients = {
        "dW": dW,
        "db": db
    }
    return params, gradients, costs

def predict(W, b, X): #https://es.stackoverflow.com/questions/569777/simplificar-operacion-matricial-y-vectorizada-en-una-sola-funcion-con-numpy
    Z = np.dot(W, X.T) + b 
    A = kernel_sigmoid(Z)
    y_prediction = (A > 0.5).astype(int)  #Cambio, y_prediction[0, i] = 1 if A[0, i] > 0.5 else 0
    return y_prediction


def model_regresion_logistico_optimizado(X_train, y_train, X_val, y_val, num_iterations=2000, learning_rate=0.5):
    dimensions = X_train.shape[1]
    W = np.zeros(shape=(1, dimensions))
    b = 0
    params, gradients, costs = optimize(W, b, X_train, y_train, num_iterations, learning_rate)
    W = params["W"]
    b = params["b"]
    y_prediction_validation = predict(W, b, X_val)
    y_prediction_train = predict(W, b, X_train)
    lista = {
        "Ajuste entrenamiento": 100 - np.mean(np.abs(y_prediction_train - y_train)) * 100,
        "Ajuste testeo": 100 - np.mean(np.abs(y_prediction_validation - y_val)) * 100,
        "Costo en proceso": costs
    }
    return lista

def medir_tiempo(func, *args, **kwargs):
    tiempos = []
    for i in range(1, 11):
        inicio = time.time()
        func(*args, **kwargs)
        fin = time.time()
        tiempos.append((i, fin - inicio))
    return tiempos

datos = pd.read_csv("Diabetes.txt")
X = np.array(datos.drop(["Outcome"], axis=1))
y = np.array(datos["Outcome"])
X = (X - np.min(X)) / (np.max(X) - np.min(X))

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.20, random_state=42)

# Modelo optimizados
tiempos_optimizado = medir_tiempo(model_regresion_logistico_optimizado, X_train, y_train, X_val, y_val, num_iterations=1000, learning_rate=0.003)

# Resultados
resultados = pd.DataFrame({
    "Run": [i for i in range(1, 11)],
    "Tiempos Optimizado": [tiempo[1] for tiempo in tiempos_optimizado]
})

print(resultados)

conn = sqlite3.connect('resultados.db')
resultados.to_sql('resultados_optimizado', conn, if_exists='replace', index=False)
conn.close()
```

```python
#https://www.freecodecamp.org/news/how-to-build-and-train-linear-and-logistic-regression-ml-models-in-python/
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import time

#codigo original
def kernel_sigmoid(z):
    return 1 / (1 + np.exp(-z))

def forward_propagation(W, b, X, y):
    data_number = X.shape[0]
    Z = np.dot(W, X.T) + b
    A = kernel_sigmoid(Z)
    cost = (-1 / data_number) * np.sum(y * np.log(A) + (1 - y) * (np.log(1 - A)))
    return A, cost

def backward_propagation(X, A, y):
    data_number = X.shape[0]
    dW = (1 / data_number) * np.dot((A - y), X)
    db = (1 / data_number) * np.sum(A - y)
    return dW, db

def optimize(W, b, X, y, num_iterations, learning_rate):
    costs = []
    for i in range(num_iterations):
        A, cost = forward_propagation(W, b, X, y)
        dW, db = backward_propagation(X, A, y)
        W = W - learning_rate * dW
        b = b - learning_rate * db
        if i % 100 == 0:
            costs.append(cost)
    params = {"W": W, "b": b}
    gradients = {"dW": dW, "db": db}
    return params, gradients, costs

def predict(W, b, X):
    data_number = X.shape[0]
    y_prediction = np.zeros((1, data_number))
    Z = np.dot(W, X.T) + b
    A = kernel_sigmoid(Z)
    for i in range(A.shape[1]):
        y_prediction[0, i] = 1 if A[0, i] > 0.5 else 0
    return y_prediction

def model_regresion_logistico(X_train, y_train, X_val, y_val, num_iterations=2000, learning_rate=0.5):
    dimensions = X.shape[1]
    W = np.zeros(shape=(1, dimensions))
    b = 0
    params, gradients, costs = optimize(W, b, X_train, y_train, num_iterations, learning_rate)
    W = params["W"]
    b = params["b"]
    y_prediction_validation = predict(W, b, X_val)
    y_prediction_train = predict(W, b, X_train)
    lista = {
        "Ajuste entrenamiento": 100 - np.mean(np.abs(y_prediction_train - y_train)) * 100,
        "Ajuste testeo": 100 - np.mean(np.abs(y_prediction_validation - y_val)) * 100,
        "Costo en proceso": costs
    }
    return lista

# https://ellibrodepython.com/tiempo-ejecucion-python, se tomo: 
#def mide_tiempo(funcion):
    #def funcion_medida(*args, **kwargs):
        #inicio = time.time()
        #c = funcion(*args, **kwargs)
        #print(time.time() - inicio)
        #return c
    #return funcion_medida
    
def medir_tiempo(func, *args, **kwargs):
    tiempos = []
    for i in range(1, 11):
        inicio = time.time()
        func(*args, **kwargs)
        fin = time.time()
        tiempos.append((i, fin - inicio))
    return tiempos

#Datos 
datos = pd.read_csv("Diabetes.txt")
X = np.array(datos.drop(["Outcome"], axis=1))
y = np.array(datos["Outcome"])
X = (X - np.min(X)) / (np.max(X) - np.min(X))

# Separa la muestra
X_train, X_val, y_train, y_val = train_test_split(
    X, 
    y,  
    random_state=None,  
    test_size=0.20  )

# Ejecucion para el dado
tiempos_custom = medir_tiempo(model_regresion_logistico, X_train, y_train, X_val, y_val, num_iterations=1000, learning_rate=0.003)

# Ejecucións para LogisticRegression 
tiempos_sklearn = medir_tiempo(LogisticRegression().fit, X_train, y_train) #ver link de arriba 

# resultados
resultados = pd.DataFrame({
    "Run": [i for i in range(1, 11)],
    "tiempos original": [tiempo[1] for tiempo in tiempos_custom],
    "tiempos sklearn": [tiempo[1] for tiempo in tiempos_sklearn]
})

print(resultados)

conn = sqlite3.connect('resultados.db')

resultados.to_sql('resultados_original_vs_sklearn', conn, if_exists='replace', index=False)
```
